# Cleaning text

```python
def clean_text (text) :
    text = "".join([ponct for ponct in text if ponct not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text
```

### remove punctuation and uppercase

```python
dataset["comment_no_ponct"] = dataset["comment"].apply(lambda x: clean_text(x.lower())) 
```

# Lemmatization

```python
def lemmatization(token_text):
    text = [wn.lemmatize(word,pos='v') for word in token_text]
    return text
```

```python
dataset["comment_lemmatized"] = dataset["comment_no_ponct"].apply(lambda x : lemmatization(x) )
```
